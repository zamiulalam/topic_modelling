```python
# create preprocess_text function
def preprocess_text(text):

    text = str(text)

    text = re.sub(r'—', '', text)

    # Remove months
    text = re.sub(r'\b(?:Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)\b', '', text)

    # Remove dates (1–31)
    text = re.sub(r'\b(?:[1-9]|[12][0-9]|3[01])\b', '', text)

    # Remove newspaper name in parentheses except (AI)
    text = re.sub(r'\s*\((?!AI\))[^)]*\)', '', text).strip()

    return text

# apply the function df

dff['Hleadz'] = dff['Hlead'].apply(preprocess_text)
dff
```

```python
docs = dff['Hleadz'].tolist()
```

```python
docs[0]
```

```python
from sentence_transformers import SentenceTransformer
embedding_model = SentenceTransformer("all-MiniLM-L6-v2")

topic_model = BERTopic(
    embedding_model=embedding_model,
    vectorizer_model=CountVectorizer(stop_words="english"),
    min_topic_size=20
    )

topics, probs = topic_model.fit_transform(docs)
```
